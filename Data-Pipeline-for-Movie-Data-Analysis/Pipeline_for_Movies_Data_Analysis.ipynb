# Requests

## 2.1 API Request

```python
import requests
import pandas as pd

# Substitua 'YOUR_API_KEY' pela sua chave de API do TMDb
api_key = 'a43023d7cabab33491d9cdf1c3c11b35'
```

## 2.2 Making the Request

```python
# URL base da API do TMDb
base_url = 'https://api.themoviedb.org/3'

# Endpoint para filmes populares
popular_movies_endpoint = f'{base_url}/movie/popular'

# Parâmetros da requisição
params = {
    'api_key': api_key,
    'language': 'en-US',  # Idioma da resposta
    'page': 1  # Página dos resultados
}

# Fazendo a requisição
response = requests.get(popular_movies_endpoint, params=params)

# Verificando o status da requisição
if response.status_code == 200:
    popular_movies = response.json()
    print(popular_movies)  # Imprime a resposta para inspecionarmos
else:
    print(f"Erro: {response.status_code}")
```

## 2.3 Extracting Data

```python
# Extracting the list of movies
movies_list = popular_movies['results']

# Creating a DataFrame from the list of movies
movies_df = pd.DataFrame(movies_list, columns=['id', 'title', 'release_date', 'vote_average', 'vote_count', 'overview', 'genre_ids'])

# Displaying the first rows of the DataFrame
movies_df.head()
```

## 3. Saving the Data

```python
# Saving the data to a CSV file
movies_df.to_csv('popular_movies.csv', index=False)
```

## Example:Displaying the top 10 rated movies

```python
import requests
import pandas as pd

# Your TMDb API key
api_key = 'a43023d7cabab33491d9cdf1c3c11b35'

# URL for top-rated movies
top_rated_url = 'https://api.themoviedb.org/3/movie/top_rated'

# Parameters for the API request
params = {
    'api_key': api_key,
    'language': 'en-US',
    'page': 1  # Fetching the first page of results
}

# Making the GET request to the API
response = requests.get(top_rated_url, params=params)

# Checking if the request was successful
if response.status_code == 200:
    top_rated_movies = response.json()
    top_rated_movies_df = pd.DataFrame(top_rated_movies['results'], columns=['title', 'release_date', 'vote_average', 'overview'])

    # Displaying the top 10 rated movies
    print(top_rated_movies_df.head(10))
else:
    print(f"Error: {response.status_code}")
```
